```python
# -*- coding: utf-8 -*-
"""Equality Constraint Using Simplex.ipynb
Automatically generated by Colaboratory.
Original file is located at
https://colab.research.google.com/drive/1XknmWIr_O9_uZlPh2njdB6d2HGdU_9oi
Name: Kajal Panigrahi
Roll No: 22
---
**Use Simplex Method for below equation:**
Max = -1x-2y
2x+y<=20
-4x+5y<=10
1x-2y<=2
-1x+5y=15
x,y>0
"""
from scipy.optimize import linprog
obj=[-1,-2]
lhs_ineq=[[2,1],[-4,5],[1,-2]]
rhs_ineq=[20,10,2]
lhs_eq=[[-1,5]]
rhs_eq=[15]
bnd=[(0,float("inf")),(0,float("inf"))]
opt=linprog(c=obj,A_ub=lhs_ineq, b_ub=rhs_ineq,A_eq=lhs_eq,
b_eq=rhs_eq,bounds=bnd,method="revised simplex")
opt
```
